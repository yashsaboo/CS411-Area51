```python
import csv
import pandas as pd
```

```python
csvFilePath = r"C:\Users\Yash\Desktop\Courses\CS411\Project Track 1\Src\Convert_CSV\Data"
csvList = ["2013new.csv", "2014new.csv", "2015new.csv", "2016new.csv", "2017new.csv", "2018new.csv", "2019new.csv"]
```

```python
csvData = pd.DataFrame()
for csvName in csvList:
    csvDataTemp = pd.read_csv(csvFilePath+"\\"+csvName)
    csvData = pd.concat([csvData,csvDataTemp])
listOfcsvDataColumns = list(csvData.columns)
csvData
```

```python
# csvData = pd.read_csv(csvFilePath+"\\"+csvName)
# listOfcsvDataColumns = list(csvData.columns)
# csvData
```

```python
# Drop NaN rows
csvData = csvData.dropna()
csvData = csvData.reset_index(drop=True)

# Drop row which has "nan" value
for col in listOfcsvDataColumns:
    csvData = csvData[~csvData[col].isin(["nan", "UNKNOWN", "unknown"])]

csvData
```

# Create CrimeType CSV

```python
#Get unique types of crime
crimeTypeSet = set()
for x in csvData["Crime Description"]:
    crimeTypeSet.add(x.upper().strip())
crimeTypeList = list(crimeTypeSet)
crimeTypeList
```

```python
#Convert the set of crime types into a list of dictionary
crimeTypeListOfDicts = []
crimeTypeStartIndex = 300
for type in crimeTypeList:
    crimeTypeDict = {}
    crimeTypeDict["crimeTypeID"] = crimeTypeStartIndex
    crimeTypeDict["type"] = type
    crimeTypeStartIndex += 1
    crimeTypeListOfDicts.append(crimeTypeDict)
crimeTypeListOfDicts
```

```python
#Flush the list of dictionary to CSV file
crimeTypeColumns = ['crimeTypeID','type']

crimeTypeCSVName = "CrimeType.csv"

try:
    with open(csvFilePath+"\\"+crimeTypeCSVName, 'w', newline="") as csvfile:
        writer = csv.DictWriter(csvfile, fieldnames=crimeTypeColumns)
        writer.writeheader()
        for data in crimeTypeListOfDicts:
            writer.writerow(data)
except IOError:
    print("I/O error")
```

# Create Crime CSV

### Add incident column

```python
crimeDF = csvData[["Incident"]]
crimeDF
```

### Add reportedAt column

```python
def convertCSVDateToSQLDateFormat(str):
    list = str.split("/")
    res = "20" + list[2] + "-" + list[1] + "-" + list[0]
    return res
convertCSVDateToSQLDateFormat("1/1/13")
```

```python
# Add reportedAt column
reportedAt = []
for date in csvData["Date reported"]:
    reportedAt.append(convertCSVDateToSQLDateFormat(date))

index = 0
for time in csvData["Time reported"]:
    reportedAt[index] += " " + time.split(" ")[0]
    index += 1
    
reportedAtDF = pd.DataFrame(reportedAt, columns = ['reportedAt'])
reportedAtDF
```

```python
#Combine it with crimeDF
crimeDF = pd.concat([crimeDF,reportedAtDF], axis=1)
crimeDF
```

### Add occuredAt column

```python
# Add occuredAt column
# print(csvData.isnull())

occurredAt = []
for date in csvData["Date occurred"]:
    print(date)
#     print(type(date))
    occurredAt.append(convertCSVDateToSQLDateFormat(date))

index = 0
for time in csvData["Time occurred"]:
    occurredAt[index] += " " + time.split(" ")[0]
    index += 1
    
occurredAtDF = pd.DataFrame(occurredAt, columns = ['occurredAt'])
occurredAtDF
```

```python
#Combine it with crimeDF
crimeDF = pd.concat([crimeDF,occurredAtDF], axis=1)
crimeDF
```

### Add Disposition column

```python
#Combine it with crimeDF
crimeDF = pd.concat([crimeDF,csvData["Disposition"]], axis=1)
crimeDF
```

### Add crimeTypeID column

```python
crimeTypeID = []
for row in csvData["Crime Description"]:
    crimeTypeStartIndex = 300
    for type in crimeTypeList:
        if(type==row.upper().strip()):
            crimeTypeID.append(crimeTypeStartIndex)
            break;
        crimeTypeStartIndex += 1

crimeTypeIDDF = pd.DataFrame(crimeTypeID, columns = ['crimeTypeID'])
crimeTypeIDDF
```

```python
#Combine it with crimeDF
crimeDF = pd.concat([crimeDF,crimeTypeIDDF], axis=1)
crimeDF
```

```python
#Rename columns
crimeDF.columns = ["incidentID", "reportedAt", "occuredAt", "disposition", "crimeTypeID"]
crimeDF
```

```python
#Flush the dataframe to CSV file

crimeCSVName = "Crime.csv"
    
crimeDF.to_csv(csvFilePath+"\\"+crimeCSVName, encoding='utf-8', index=False)
```
